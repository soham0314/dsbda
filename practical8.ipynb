```python
#The seaborn library can be downloaded in a couple of ways. If you are using pip installer fo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#All you have to do is use the load_dataset function and pass it the name of the dataset.
dataset = sns.load_dataset('titanic')
dataset.head()
```

```python
#The Dist Plot The distplot() shows the histogram distribution of data for a single column. T
sns.distplot(dataset['fare'])
sns.distplot(dataset['fare'], kde=False)
```

```python
#You can also pass the value for the bins parameter in order to see more or less details in t
#Take a look at he following script:
sns.jointplot(x='age', y='fare', data=dataset)
```

```python
sns.distplot(dataset['fare'], kde=False, bins=10)
```

```python
sns.jointplot(x='age', y='fare', data=dataset, kind='hex')
```

```python
#The Pair Plot The paitplot() is a type of distribution plot that basically plots a joint plo
dataset = dataset.dropna()
```

```python
#The rugplot() is used to draw small bars along x-axis for each point in the dataset.To plot
sns.rugplot(dataset['fare'])
```
